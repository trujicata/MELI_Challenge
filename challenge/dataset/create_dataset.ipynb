# Create Train and Test Dataset

Since the creation of the dataset takes a lot of time because of the large amount of data, the API for the dollar/peso conversion and the encoder, we will create the datasets in this notebook and store it

```python
import pandas as pd
import pickle
from challenge.new_or_used import build_dataset
from challenge.dataset.preprocess import preprocess_whole_dataset


X_train, y_train, X_test, y_test = build_dataset()


df = pd.DataFrame(X_train)
df.head()
```

```python
X_df = preprocess_whole_dataset(df, use_encoders=True)
X_df.head()
```

```python
for col in X_df.columns:
    try:
        if X_df[col].dtype == "object":
            X_df[col] = X_df[col].astype("category")
    except:
        print(col)
X_df.dtypes
```

```python
# Save the processed dataset into the data folder
with open("../data/train_dataset.pkl", "wb") as f:
    pickle.dump(X_df, f)
```

```python
X_df_test = pd.DataFrame(X_test)
X_df_test = preprocess_whole_dataset(X_df_test, use_encoders=True)
for col in X_df_test.columns:
    try:
        if X_df_test[col].dtype == "object":
            X_df_test[col] = X_df_test[col].astype("category")
    except:
        print(col)
X_df_test.dtypes
```

```python
with open("../data/test_dataset.pkl", "wb") as f:
    pickle.dump(X_df_test, f)
```
